```python
def point(grade):
    if grade>=90:
        return 'A+'
    elif grade<90 and grade>=85:
        return 'A'
    elif grade<85 and grade>=80:
        return 'A-'
    elif grade<80 and grade>=77:
        return 'B+'
    elif grade<77 and grade>=73:
        return 'B'
    elif grade<73 and grade>=70:
        return 'B-'
    elif grade<70 and grade>=67:
        return 'C+'
    elif grade<67 and grade>=63:
        return 'C'
    elif grade<63 and grade>=60:
        return 'C-'
    else:
        return 'D'
```

```python
point(df.iloc[0]['總成績'])
```

```python
gpa_four_point_three = []
for i in df['總成績'][0:len(df)]:
    gpa_four_point_three.append(point(i))
```

```python
df['等第制'] = gpa_four_point_three
```
